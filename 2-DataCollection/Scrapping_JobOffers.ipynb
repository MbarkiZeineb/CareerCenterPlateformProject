```python
# Access to the Jobs button and click it
# Go to search results directly via link
driver.get("https://fabskill.com/candidate/search")
time.sleep(1)
```

```python
links=[]
```

```python
try: 
    jobs_block = driver.find_element(By.CLASS_NAME,"row ng-tns-c135-3")
    time.sleep(2)
    for job in jobs_block:
        all_links = job.find_elements(By.TAG_NAME,"a")
        for a in all_links:
            if str(a.get_attribute('href')).startswith("https://fabskill.com/candidate/search") and a.get_attribute('href') not in links: 
                links.append(a.get_attribute('href'))
                else:
                    pass
            # scroll down for each job element
            driver.execute_script("arguments[0].scrollIntoView();", job)
        
        print(f'Collecting the links in the page: {page-1}')
        # go to next page:
        driver.find_element("xpath",f"//button[@aria-label='Page {page}']").click()
        time.sleep(3)
except:
    pass
print('Found ' + str(len(links)) + ' links for job offers')
```

```python
job_titles = []
company_names = []
company_locations = []
work_methods = []
post_dates = []
work_times = [] 
job_desc = []
```

```python
i = 0
j = 1
```

```python
print('Visiting the links and collecting information just started.')
for i in range(len(links)):
    try:
        driver.get(links[i])
        i=i+1
        print(i)
        time.sleep(2)
        # Click See more.
        driver.find_element(By.CLASS_NAME,"artdeco-card__actions").click()
        time.sleep(2)
    except:
        pass
    
    # Find the general information of the job offers
    contents = driver.find_elements(By.CLASS_NAME,'p5')
    for content in contents:
        try:
            job_tit =content.find_element(By.TAG_NAME,"h1").text
            job_titles.append(job_tit)
            company_n=content.find_element(By.CLASS_NAME,"jobs-unified-top-card__company-name").text
            company_names.append(company_n)
            company_loc=content.find_element(By.CLASS_NAME,"jobs-unified-top-card__bullet").text
            company_locations.append(company_loc)
            company_work=content.find_element(By.CLASS_NAME,"jobs-unified-top-card__workplace-type").text
            work_methods.append(company_work)
            company_date=content.find_element(By.CLASS_NAME,"jobs-unified-top-card__posted-date").text
            post_dates.append(company_date)
            company_time=content.find_element(By.CLASS_NAME,"jobs-unified-top-card__job-insight").text
            work_times.append(company_time)
            print(f'Scraping the Job Offer {j} DONE.')
            j+= 1
        except:
            pass
        time.sleep(2)
        # Scraping the job description
    job_description = driver.find_elements(By.CLASS_NAME,'jobs-description__content')
    for description in job_description:
        job_text = description.find_element(By.CLASS_NAME,"jobs-box__html-content").text
        job_desc.append(job_text)
        print(f'Scraping the Job Offer {j}')
        time.sleep(2)
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(job_titles,company_names,company_locations,work_methods,post_dates,work_times,job_desc)),columns =['job_title', 'company_name','company_location','work_method','post_date','work_time','job_desc'])
```
