```python
import numpy as np
import re
import pandas as pd
import pdfplumber
import PyPDF2
```

```python
#Extract text from pdf 
def pdfextraction(path) :
    all_text = ""
    name=""
    with pdfplumber.open(path) as pdf:
        for pdf_page in pdf.pages :
            name=pdf.stream.name
            single_page_text = pdf_page.extract_text()
            all_text= all_text+'\n'+single_page_text
    return all_text,name
#Extract Description from text
def Pattern2keys(text,df,key1,key2):
    pattern = r""+key1+"(.*?)"+key2
    captured_texts = re.findall(pattern,text,re.DOTALL)
    df[key1] = captured_texts
def duree(text,df):
    pattern = r"Durée\s*:\s*(.*?)\w\d"
    captured_texts = re.findall(pattern,text,re.DOTALL)
    df["Duree"] = captured_texts
def extract_text_from_pdf(pdf_file):
    with open(pdf_file, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)
        text = ''
        for page in range(num_pages):
            page_obj = pdf_reader.pages[page]
            text += page_obj.extract_text()
        return text
```
