```python
Data1=data
Localisation=Data1[['Location']]
Localisation
```

```python
DimLocalisation= Dimension( name='DimLocalisation',
                         key='Localisation_PK',               
                         attributes=['Location'], 
                         lookupatts=(['Location']),
                         idfinder=None,
                         defaultidvalue=-1
                        )
```

```python
dic={}
for j in Localisation.index: 
    for i in Localisation:
        dic[i]=Localisation[i][j]
    
    
      
    key = DimLocalisation.lookup(dic)
    print(key)
    if key==-1:
        DimLocalisation.insert(dic)
    else:
        
        DimLocalisation.ensure(dic)
    
conn2.commit()
```

```python
factOffers = FactTable(name='fact_Offers',
                       keyrefs=['Offer_PK', 'Location_FK','Field_FK','Type_FK'])
```

```python
dic={}
for j in data.index:
    for i in data:
        dic[i]=data[i][j]
        dic['Offer_PK'] = j+1  # Fix indentation and use j instead of i
        
    dic['Location_FK'] = DimLocalisation.lookup(dic)
    dic['Field_FK'] = DimField.lookup(dic)
    dic['Type_FK'] = DimType.lookup(dic)

    if dic['Location_FK']!=-1 and dic['Field_FK']!=-1 and dic['Type_FK']!=-1:
        factOffers.ensure(dic)
        conn2.commit()

    else:
        pass

        

# Join fact table with location and type dimension tables to get their names
cursor.execute('''
SELECT f.Offer_PK, l.Location, d.Field, t.Type
FROM DW.dbo.fact_Offers f
INNER JOIN DW.dbo.DimLocalisation l ON f.Location_FK = l.Localisation_PK
INNER JOIN DW.dbo.DimField d ON f.Field_FK = d.Field_PK
INNER JOIN DW.dbo.DimType t ON f.Type_FK = t.Type_PK
''')
result = cursor.fetchall()
print(result)

conn2.commit()

  
```
