```python
nlp = spacy.load('fr_core_news_sm')
def add_newruler_to_pipeline(skill_pattern_path):
    '''Reads in all created patterns from a JSONL file and adds it to the pipeline after PARSER and before NER'''
    
    #new_ruler = EntityRuler(nlp).from_disk(skill_pattern_path)
    ruler=nlp.add_pipe('entity_ruler', after='parser')
    ruler.from_disk(skill_pattern_path)  # loads patterns only
    return ruler
nlp.add_pipe = add_newruler_to_pipeline("c:/users/Iheb/Desktop/pi/pattern/skill_patterns.jsonl")
```

```python
from os import listdir
from os.path import isfile, join
mypath = r"C:/users/Iheb/Desktop/PI/cv/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles
```

```python
def create_skill_set(doc):
    '''Create a set of the extracted skill entities of a doc'''    
    return [ent.label_.upper()[6:] for ent in doc.ents if 'skill' in ent.label_.lower()]
def extract_text_from_pdf(file):
    '''Opens and reads in a PDF file from path'''
    
    fileReader = PyPDF2.PdfReader(open(file,'rb'))
    page_count = len(fileReader.pages)
    text = [fileReader.pages[i].extract_text() for i in range(page_count)]
    return str(text).replace("\\n", " ")
skill = []
for files in range(len(onlyfiles)):
    CV_text = extract_text_from_pdf('C:/users/Iheb/Desktop/PI/cv/'+onlyfiles[files])
    doc = nlp(CV_text)
    
    cv_skills = create_skill_set(doc)
    skill.append(cv_skills)
    
skill
```
