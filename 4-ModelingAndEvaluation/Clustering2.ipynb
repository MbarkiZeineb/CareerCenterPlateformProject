**KMeans** **Algorithm**

We used KMeans clustering algorithm to cluster the job titles and we used **Elbow** **method** to find the optimal number of clusters

```python
def cluster_text(text):
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(text)

    import matplotlib.pyplot as plt
    from sklearn.cluster import KMeans
    Sum_of_squared_distances = []
    K = range(2,20)
    for k in K:
       km = KMeans(n_clusters=k, max_iter=200, n_init=10)
       km = km.fit(X)
       Sum_of_squared_distances.append(km.inertia_)
    plt.plot(K, Sum_of_squared_distances, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Sum_of_squared_distances')
    plt.title('Elbow Method For Optimal k')
    plt.show()

    print('How many clusters do you want to use?')
    true_k = int(input())
    model = KMeans(n_clusters=true_k, init='k-means++', max_iter=200, n_init=10)
    model.fit(X)

    labels=model.labels_
    clusters=pd.DataFrame(list(zip(text,labels)),columns=['title','cluster'])
    #print(clusters.sort_values(by=['cluster']))

    for i in range(true_k):
        print(clusters[clusters['cluster'] == i])
        
    return clusters
```

```python
import spacy
def add_newruler_to_pipeline(skill_pattern_path):
    '''Reads in all created patterns from a JSONL file and adds it to the pipeline after PARSER and before NER'''
    
    #new_ruler = EntityRuler(nlp).from_disk(skill_pattern_path)
    #ruler=nlp.add_pipe('entity_ruler', after='parser')
    ruler=nlp.add_pipe('entity_ruler', after='parser')
    ruler.from_disk(skill_pattern_path)  # loads patterns only
    return ruler
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe = add_newruler_to_pipeline("/content/skill_patterns.jsonl")
skillsfinal=[]
for text in data['jobDescription'] :
  new_list=[]
  skills_list = []
  doc=nlp(text)
  for ent in doc.ents:
    if ent.label_.startswith("SKILL") or ent.label_=="ORG" :
      skills_list.append(ent.text)
  for value in skills_list:
        if value not in new_list:
            new_list.append(value)
  skillsfinal.append(str(new_list))
data['Skills_Description']=skillsfinal
```

```python
# Group the data by cluster
# Create an empty report dataframe
import ast
report_df = pd.DataFrame(columns=['Field', 'Technologies','Company','Location','Title'])
```

```python
for name, group in grouped:
   my_list = []
   location = group['Country'].apply(str).unique()
   company = group['companyName'].apply(str).unique()
   JobTitle = group['jobTitle'].apply(str).unique()
   technologies_list = group['Skills_Description'].apply(str).unique()
   my_list = technologies_list
   flat_list = list(set([item for sublist in [ast.literal_eval(x) for x in my_list] for item in sublist]))
   report_df = report_df.append({'Field': name , 'Technologies':flat_list , 'Company': company,'Location':location,'Title': JobTitle }, ignore_index=True)
```
